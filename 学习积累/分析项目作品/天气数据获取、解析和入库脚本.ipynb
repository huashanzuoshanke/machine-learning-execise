# json 解析

```python
weather_info = pd.DataFrame() #创建一个空数据框
for i in range(len(df.index.to_list())):  #遍历每条json
    data = df['results'][i] #results字段中，每一条数据都是list，每一条数据都是list类型，可直接传入json_normalize方法
    analyse = pd.json_normalize(data,'daily',[['last_update'],['location','country']
                                  ,['location','name']
                                  ,['location','id']
                                  ,['location','path']
                                  ,['location','timezone']
                                  ,['location','timezone_offset']])
    #data:将一串list传入json_normalize
    #往后，以列表嵌套字典键的形式，作为我们需要规范的字段，
    #如果键对应的值不是字典，如last_update，则直接嵌套一个列表即可：['last_update']；
    #如果键对应的值是字典，如location，我们以['location','country']的形式
    #以上内容需放在同一个[]内
    #注意，daily自身就为表嵌套字典键的形式，不需要加[]
    weather_info = pd.concat([weather_info,analyse])
```
